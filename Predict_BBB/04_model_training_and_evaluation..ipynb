```python
import numpy as np
def remove_highly_correlated_features(df, threshold=0.85, method='spearman'):

    corr_matrix = df.corr(method=method).abs()

    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

    df_reduced = df.drop(columns=to_drop)

    return df_reduced, to_drop
```

```python
X_uncorrelated, dropped_features = remove_highly_correlated_features(X_desc, threshold=0.85, method='spearman')
print("Dropped features:", dropped_features)
```

```python
print(f"Original features (after variance filter): {X_desc.shape[1]}")
print(f"Remaining features (after correlation filter): {X_uncorrelated.shape[1]}")
print(f"Features removed due to high correlation: {len(dropped_features)}")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_uncorrelated, y, test_size=0.2, random_state=42, stratify=y)
```

```python
X_train.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score, matthews_corrcoef, confusion_matrix, classification_report
from catboost import CatBoostClassifier
```

```python
from collections import Counter

counter = Counter(y_train)
total = sum(counter.values())
class_weights = [total / counter[0], total / counter[1]]
```

```python
class_weights
```

```python
model = CatBoostClassifier(
    iterations=2000,
    depth=6,
    learning_rate=0.03,
    loss_function='Logloss',
    eval_metric='AUC',
    random_seed=42,
    l2_leaf_reg=3,
    class_weights=class_weights,
    od_wait=50,
    verbose=200
)

model.fit(X_train, y_train, eval_set=(X_test, y_test), use_best_model=True)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, matthews_corrcoef, confusion_matrix, classification_report

y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]

accuracy = accuracy_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred_proba)
mcc = matthews_corrcoef(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f" Accuracy: {accuracy:.4f}")
print(f" AUC: {auc:.4f}")
print(f" MCC: {mcc:.4f}")
print("\nConfusion Matrix:")
print(cm)
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['-BBB', '+BBB']))
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.svm import SVC
```

```python
from sklearn.metrics import (matthews_corrcoef, f1_score, roc_auc_score,classification_report)
```

```python
models = {
    "Random Forest": RandomForestClassifier(class_weight='balanced', random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='auc', random_state=42),
    "Extra Trees": ExtraTreesClassifier(class_weight='balanced', random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=5000, class_weight='balanced', random_state=42),
    "SVC": SVC(probability=True, class_weight='balanced', random_state=42)
}
```

```python
def evaluate_model(model, name, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_prob = model.predict_proba(X_test)[:, 1]

    thresholds = np.linspace(0, 1, 100)
    mcc_scores = [matthews_corrcoef(y_test, (y_prob >= t).astype(int)) for t in thresholds]
    best_idx = np.argmax(mcc_scores)
    best_threshold = thresholds[best_idx]

    y_pred_default = (y_prob >= 0.5).astype(int)
    auc_default = roc_auc_score(y_test, y_prob)
    mcc_default = matthews_corrcoef(y_test, y_pred_default)
    f1_default = f1_score(y_test, y_pred_default)
    acc_default = accuracy_score(y_test, y_pred_default)

    y_pred_best = (y_prob >= best_threshold).astype(int)
    auc_best = auc_default  
    mcc_best = mcc_scores[best_idx]
    f1_best = f1_score(y_test, y_pred_best)
    acc_best = accuracy_score(y_test, y_pred_best)

    print(f"\n{name} Results")
    print(" Default Threshold (0.5) ")
    print(f" AUC: {auc_default:.4f}")
    print(f" Accuracy: {acc_default:.4f}")
    print(f" MCC: {mcc_default:.4f}")
    print(f" F1 Score: {f1_default:.4f}")
    print(" Classification Report:")
    print(classification_report(y_test, y_pred_default, target_names=["-BBB", "+BBB"]))

    print(" Optimized Threshold")
    print(f" Best Threshold: {best_threshold:.4f}")
    print(f" Accuracy: {acc_best:.4f}")
    print(f" MCC: {mcc_best:.4f}")
    print(f" F1 Score: {f1_best:.4f}")
    print(" Classification Report:")
    print(classification_report(y_test, y_pred_best, target_names=["-BBB", "+BBB"]))
    print("-" * 60)

    return {
        'model': name,
        'auc': auc_best,
        'accuracy': acc_best,
        'mcc': mcc_best,
        'f1': f1_best,
        'best_threshold': best_threshold
    }
```

```python
results = []

for name, model in models.items():
    result = evaluate_model(model, name, X_train, X_test, y_train, y_test)
    results.append(result)

df_results = pd.DataFrame(results)
print("Summary Results")
print(df_results)
```
