## Median imputation and removal of low-variance features (threshold=0.01)

```python
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import VarianceThreshold
```

```python
original_columns = X_desc.columns
original_index = X_desc.index
```

```python
imputer = SimpleImputer(strategy='median')  
X_imputed = imputer.fit_transform(X_desc)
```

```python
selector = VarianceThreshold(threshold=0.01)
X_high_variance = selector.fit_transform(X_imputed)
```

```python
selected_features = original_columns[selector.get_support()]
```

```python
X_filtered = pd.DataFrame(X_high_variance, columns=selected_features, index=original_index)
```

```python
print(f"Original features: {X_desc.shape[1]}")
print(f"Remaining features: {X_filtered.shape[1]}")
print(f"Removed features: {X_desc.shape[1] - X_filtered.shape[1]}")
```

```python
X_filtered.head(1)
```

### Caps extreme values per feature to [1st, 99th] percentiles
### Keeps all rows and reduces outlier impact

```python
def winsorize_features(df, features, lower=0.01, upper=0.99):    
    df_winsor = df.copy()
    for col in features:
        low_val = df_winsor[col].quantile(lower)
        high_val = df_winsor[col].quantile(upper)
        df_winsor[col] = df_winsor[col].clip(lower=low_val, upper=high_val)
    return df_winsor
```

```python
X_filtered = winsorize_features(X_filtered, X_filtered.columns)

print(f" Winsorization complete. Shape: {X_filtered.shape}")
```

```python
top_var_features = X_filtered.var().sort_values(ascending=False).head(10).index.tolist()
print("Top variable features:", top_var_features)
```

## Applying Distance Correlation

```python
import dcor
from tqdm import tqdm 

results = []

for col in tqdm(X_filtered.columns, desc="Calculating Distance Correlation"):
    dcor_val = dcor.distance_correlation(X_filtered[col].values, y.values)
    results.append((col, dcor_val))

dcor_df = pd.DataFrame(results, columns=["Feature", "DistanceCorr"])
dcor_df = dcor_df.sort_values(by="DistanceCorr", ascending=False)
```

```python
top_feats = dcor_df[dcor_df['DistanceCorr'] > 0.25]['Feature'].tolist()

X_top_dcor = X_filtered[top_feats]
```

```python
import numpy as np
def remove_highly_correlated_features(df, threshold=0.85, method='spearman'):

    corr_matrix = df.corr(method=method).abs()

    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

    df_reduced = df.drop(columns=to_drop)

    return df_reduced, to_drop
```

```python
X_uncorrelated, dropped_features = remove_highly_correlated_features(X_top_dcor, threshold=0.85, method='spearman')
print("Dropped features:", dropped_features)
```
