```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('dcorr_filtered_BBB_0.25.csv', low_memory=False)
```

```python
df.shape
```

```python
X_desc = df.drop(columns=['SMILES', 'BBB_Class'])
y = df['BBB_Class']
```

```python
X_desc.head(1)
```

```python
import numpy as np
def remove_highly_correlated_features(df, threshold=0.85, method='spearman'):

    corr_matrix = df.corr(method=method).abs()

    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

    df_reduced = df.drop(columns=to_drop)

    return df_reduced, to_drop
```

```python
X_uncorrelated, dropped_features = remove_highly_correlated_features(X_desc, threshold=0.85, method='spearman')
print("Dropped features:", dropped_features)
```

```python
print(f"Original features (after variance filter): {X_desc.shape[1]}")
print(f"Remaining features (after correlation filter): {X_uncorrelated.shape[1]}")
print(f"Features removed due to high correlation: {len(dropped_features)}")
```

```python
X_uncorrelated.head(2)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_uncorrelated, y, test_size=0.2, random_state=42, stratify=y)
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import (
    accuracy_score, roc_auc_score, confusion_matrix,
    classification_report, matthews_corrcoef
)
```

```python
model = LGBMClassifier(random_state=42, class_weight='balanced')
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]
```

```python
print(f" Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f" AUC: {roc_auc_score(y_test, y_prob):.4f}")
print(f" MCC: {matthews_corrcoef(y_test, y_pred):.4f}")
print(" Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\n Classification Report:\n", classification_report(y_test, y_pred, target_names=["-BBB", "+BBB"]))
```

```python
import numpy as np
from sklearn.metrics import f1_score, precision_score, recall_score
```

```python
thresholds = np.linspace(0, 1, 100)
mcc_scores, f1_scores, precisions, recalls = [], [], [], []

for thresh in thresholds:
    y_pred_thresh = (y_prob >= thresh).astype(int)
    mcc_scores.append(matthews_corrcoef(y_test, y_pred_thresh))
    f1_scores.append(f1_score(y_test, y_pred_thresh))
    precisions.append(precision_score(y_test, y_pred_thresh))
    recalls.append(recall_score(y_test, y_pred_thresh))

best_idx = np.argmax(mcc_scores)
best_threshold = thresholds[best_idx]
```

```python
print(f" Best Threshold: {best_threshold:.4f}")
print(f" Best MCC: {mcc_scores[best_idx]:.4f}")
print(f" F1 Score: {f1_scores[best_idx]:.4f}")
print(f" Precision: {precisions[best_idx]:.4f}")
print(f" Recall: {recalls[best_idx]:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(thresholds, mcc_scores, label='MCC', color='blue')
plt.plot(thresholds, f1_scores, label='F1 Score', color='orange', linestyle='--')
plt.axvline(best_threshold, color='red', linestyle=':', label=f'Best Threshold = {best_threshold:.3f}')
plt.title("MCC and F1 vs Threshold")
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from rdkit import Chem
from mordred import Calculator, descriptors
```

```python
external_smiles = [
    # Known permeable
    "CCO",                                       # Ethanol (permeable, small & lipophilic) [Pardridge, 2012]
    "CC(=O)O",                                   # Acetic acid (permeable) [Pardridge, 2012]
    "c1ccccc1",                                  # Benzene (permeable) [Oldendorf, 1971]
    "C1[C@H]2[C@H](CN1)OC3=C(O2)C=CC=C3F",       # Fluparoxan (permeable α2 antagonist) [Hudson et al., 2005]
    "CN1CCC(CC1)C2=CC=CC=C2O",                   # Serotonin (permeable but actively transported) [Pardridge, 1998]
    "CC(C)NCCC1=CC2=C(C=C1)OCO2",                # Donepezil (permeable, Alzheimer's drug) [Sugimoto et al., 2000]
    
    # Known NON-permeable
    "C(C(=O)O)N",                                # Glycine (polar amino acid, poor passive BBB penetration) [Pardridge, 1977]
    "C1=CC(=CC=C1C(=O)O)O",                      # Salicylic acid (polar, low BBB penetration) [Banks, 2009]
    "C1=CC(=CC=C1C(=O)O)C(=O)O",                 # Phthalic acid (very polar, non-permeable) [Banks, 2009]
    "C1=CC(=CC=C1O)O",                           # Catechol (hydrophilic, low BBB penetration) [Pardridge, 1998]
    "NC(=O)C(CC(=O)O)N",                         # Aspartic acid (non-permeable) [Pardridge, 1977]
]
```

```python
#  Compute Mordred descriptors
calc = Calculator(descriptors, ignore_3D=True)
mols = [Chem.MolFromSmiles(smi) for smi in external_smiles]
df_external = calc.pandas(mols)
```

```python
df_external.shape
```

```python
for col in ["Lipinski", "GhoseFilter"]:
    if col in df_external.columns:
        df_external[col] = df_external[col].astype(int)
```

```python
df_external = df_external.apply(pd.to_numeric, errors='coerce')
df_external = df_external.fillna(df_external.median())
```

```python
df_external = df_external[X_train.columns]
```

```python
y_prob_ext = model.predict_proba(df_external)[:, 1]
threshold = 0.5228
y_pred_ext = (y_prob_ext >= threshold).astype(int)
```

```python
results = pd.DataFrame({
    "SMILES": external_smiles,
    "Predicted_Probability": y_prob_ext,
    "Predicted_Label": y_pred_ext
})

print(results)
```
